```python
outliers=[]
def detect_outliers(ds):
    threshold=3
    mean=np.mean(ds)
    std=np.std(ds)

    for i in ds:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)

    return outliers
```

```python
detect_outliers(ds)
```

```python
#IQR

ds=sorted(ds)
```

```python
q1,q3=np.percentile(ds,[25,75])
```

```python
print(q1,q3)
```

```python
iqr=q3-q1
print(iqr)
```

```python
lower_fence=q1-(1.5*iqr)
higher_fence=q3+(1.5*iqr)

print(lower_fence,higher_fence)
```
